```python
import pandas as pd
import cutecharts.charts as ctc
import matplotlib.pyplot as plt
```

```python
# import csv file
data = pd.read_csv('severe-storms.csv')
data
```

```python
def create_line_chart(data):
    # Extracting the required data
    years = data['Year'].tolist()
    storm_counts = data['Severe Storm Count'].tolist()
    
    # Creating the line chart
    chart = ctc.Line("Severe Storm Count Over Years", width='900px', height='200px')
    chart.set_options(labels=years, x_label="Year", y_label="Severe Storm Count", colors=['#9B3192'])
    chart.add_series("Severe Storm Count", storm_counts)
    
    # Rendering the chart to an HTML file
    chart.render('severe_storm_count_over_years.html')

create_line_chart(data)

```

```python
def create_bar_chart(data):
    # Extracting the required data
    years = data['Year'].astype(str).tolist()  # Convert years to string for cutecharts
    storm_costs = data['Severe Storm Cost'].tolist()
    
    # Creating the bar chart
    chart = ctc.Bar("Severe Storm Costs Over Years (estimates)", width='1000px', height='200px')
    chart.set_options(labels=years, x_label="Year", 
                      y_label="Severe Storm Cost ($ Billion), CPI Adjusted",
                      colors=['#FFF1C9', '#FFF1C9', '#FFF1C9', '#FFF1C9', '#FFF1C9', '#FFF1C9', '#FFF1C9', 
                                '#F7B7A3', '#F7B7A3', '#F7B7A3', '#F7B7A3', '#F7B7A3', '#F7B7A3', '#F7B7A3',
                                '#EA5F89', '#EA5F89', '#EA5F89', '#EA5F89', '#EA5F89', '#EA5F89',
                                '#9B3192','#9B3192', '#57167E'])
    chart.add_series("Severe Storm Cost", storm_costs)
    
    
    # Rendering the chart to an HTML file
    chart.render('severe_storm_costs.html')

# Run the function
create_bar_chart(data)
```
